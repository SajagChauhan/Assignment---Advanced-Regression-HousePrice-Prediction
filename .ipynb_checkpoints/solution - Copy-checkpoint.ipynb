```python
# Dropping 'Id' as it does not show ant relevant significance for prediction
house=house.drop(['Id'], axis = 1)
```

```python
# Deriving new Variables
# Creating a variable 'Remodelled' to see if the house get Remodelled
def if_remodelled(data):
    if data['YearBuilt'] == data['YearRemodAdd']:
        return 0
    elif data['YearBuilt'] < data['YearRemodAdd']:
        return 1
    else:
        return "NaN"
house['Remodelled'] = house.apply(if_remodelled, axis=1)
house['Remodelled'].isnull().sum()   
```

```python
sns.scatterplot(x=house.Remodelled,y=house.SalePrice)
plt.show()
```

```python
# Creating a variable 'Age' which determines the age from Yearbuilt to YearSold or from YearRemodelled to YearSold 
def ageofhouse(data):
    if data['YearBuilt'] == data['YearRemodAdd']:
        return (data['YrSold']-data['YearBuilt'])
    elif data['YrSold'] > data['YearRemodAdd']:
        return (data['YrSold']-data['YearRemodAdd'])
    else:
        return (data['YearRemodAdd']-data['YrSold'])
house['Age'] = house.apply(ageofhouse, axis=1)
house.head()
```

```python
sns.scatterplot(house.Age,house.SalePrice)
plt.show()
# It shows that Age has strong relation with SalePrice.So dropping the YearBuilt, YearRemodAdd, YrSold will be good for our model. 
```

## Data Preparation

```python
house_categoric
```

```python
# Encoding the labels
#Converting ordered categorical variables into numeric type as that will be good for our model.
Lot={
    'Reg':3,
    'IR1':2,
    'IR2':1,
    'IR3':0
}
BsmtQual1={
    'Ex':5,
    'Gd':4,
    'TA':3,
    'Fa':2,
    'Po':1,
    'No_Basement':0
}
house['LotShape']=house['LotShape'].replace(Lot)
house['BsmtQual']=house['BsmtQual'].replace(BsmtQual1)
house['BsmtExposure']=house['BsmtExposure'].map({'Gd':4,'Av':3,'Mn':2,'No':1,'No_Basement':0})
house['BsmtFinType1']=house['BsmtFinType1'].map({'GLQ':6,'ALQ':5,'BLQ':4,'Rec':3,'LwQ':2,'Unf':1,'No_Basement':0})
house['HeatingQC']=house['HeatingQC'].map({'Ex':4,'Gd':3,'TA':2,'Fa':1,'Po':0})
house['KitchenQual']=house['KitchenQual'].map({'Ex':4,'Gd':3,'TA':2,'Fa':1,'Po':0})
house['GarageFinish']=house['GarageFinish'].map({'Fin':3,'RFn':2,'Unf':1,'No_Garage':0})
```

```python
#Except the above Categorical Variables, We will create dummy  variables for remaining categorical variables.
house_categoric_unordered = house.select_dtypes(include=['object'])
for i in house_categoric_unordered:
    dummy=pd.get_dummies(house[i],prefix=i,drop_first=True)   
    house=pd.concat([house,dummy],axis=1)
    house.drop(i,axis=1,inplace=True)   
```
